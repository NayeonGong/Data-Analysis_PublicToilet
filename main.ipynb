```python
#[장애인화장실] 필드의 value 전처리

import re

toilets["장애인화장실"] = toilets["장애인화장실"].fillna(0)

def disabled_num(x):
        try:
            x = [''.join(re.findall("\d+", item)) for item in x]
        except TypeError:
            x = x
        return x
toilets["장애인화장실"]= toilets["장애인화장실"].apply(disabled_num)

def disabled_num2(x):
    if x == 0 :
        return x
    else:
        return 1


toilets["장애인화장실"] = toilets["장애인화장실"].apply(disabled_num2)
toilets["장애인화장실"]
```
